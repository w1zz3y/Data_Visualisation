### DOB Column

```python
print("Number of missing values in the 'DOB' column:")
print(Student_Profiles['DOB'].isna().sum())
print('\n')
DOBs = Student_Profiles['DOB']

# Count the occurrences of each value
DOB_count = DOBs.value_counts()

# Print the common values
print("Common values in the 'DOB' column:")
print(DOB_count)
```

```python
# Some of the data in the DOB column are in the wrong format

# Function to convert date formats
def convert_date_format(date_str):
    for fmt in ('%d-%b-%Y', '%d-%m-%Y', '%d/%m/%Y'):
        try:
            return pd.to_datetime(date_str, format=fmt).strftime('%d/%m/%Y')
        except ValueError:
            continue

# Apply the function to the 'DOB' column
Student_Profiles['DOB'] = Student_Profiles['DOB'].apply(convert_date_format)

```
